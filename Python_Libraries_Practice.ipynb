# Plotting Square function

 Square Plot for some random values generated using linspace method. Also I have used Linestyle in this.

```python
%matplotlib inline
x=np.linspace(0,35,55)
y=x**2
print(x)
print(y)
plt.plot(x,y,color='green', linewidth=4, linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Square plot for some random values')
```

```python
%matplotlib inline
x=np.linspace(0,108,20)
y=x**3
print(x)
print(y)
plt.plot(x,y,color='blue', linewidth=7, linestyle='solid',alpha=0.5,marker='o',markerfacecolor='hotpink',markersize=9,markeredgecolor='red',markeredgewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic plot for some random values')
```

# Plotting Using Subplot

```python
fig=plt.figure(figsize=(50,5))
plt.subplot(1,10,1)
plt.plot(x,y,'b')
plt.title("Plot-1")
plt.subplot(1,10,2)
plt.plot(x**2,y**3,'red')
plt.title("Plot-2")
plt.subplot(1,10,3)
plt.plot(x**4,y,'green')
plt.title("Plot-3")
plt.subplot(1,10,4)
plt.plot(x**11,y*8,'Purple')
plt.title("Plot-4")
```

```python
x=np.linspace(0,450,50)
y=x**2
fig=plt.figure(figsize=(50,5))
plt.subplot(1,10,1)
plt.plot(x,y,'b')
plt.title("Plot-1")
plt.subplot(1,10,2)
plt.plot(x**2,y**3,'red')
plt.title("Plot-2")
plt.subplot(1,10,3)
plt.plot(x**4,y,'green')
plt.title("Plot-3")
```

# Creating one plot inside another plot

```python
fig = plt.figure()
fig = plt.figure(figsize=(10,3))
fig1=fig.add_axes([0.01,0.05,2,3])
fig2=fig.add_axes([0.5,2,0.5,0.8])
fig1.plot(x,y)
fig1.set_title('Big_Plot')
fig1.set_xlabel('x')
fig1.set_ylabel('y')
fig2.plot(x*2,y**3)
fig2.set_title('small_Plot')
fig2.set_xlabel('x')
fig2.set_ylabel('y')
```

# Using plt.tight_layout()

```python
fig,axes=plt.subplots(nrows=2, ncols=2, figsize=(10,8),dpi=100)
#Plot-1
axes[0,0].plot(x,y)
axes[0,0].set_title('Plot-1')
axes[0,0].set_xlabel("Time")
axes[0,0].set_ylabel('Production 1')
#Plot-2

axes[0,1].plot(y,x)
axes[0,1].set_title('Plot-2')
axes[0,1].set_xlabel("Time")
axes[0,1].set_ylabel('Production 2')
#Plot-3

axes[1,0].plot(4*y,x**3)
axes[1,0].set_title('Plot-3')
axes[1,0].set_xlabel("Time")
axes[1,0].set_ylabel('Production 3')
#Plot-4

axes[1,1].plot(x,y**6)
axes[1,1].set_title('Plot-4')
axes[1,1].set_xlabel("Time")
axes[1,1].set_ylabel('Production 4')
plt.tight_layout()
```

# Cubic Function Plot using values assigned by for loop

```python
t=[]
y=[]
for i in range(-100,100):
    t.append(i)
for j in t:
    y.append(j**3)
print(t)
print(y)
plt.plot(t,y, color='green', linewidth=4, linestyle='--')
plt.xlabel('time')
plt.ylabel('y')
plt.title('Cubic plot for some random values')
```
